```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm
from scipy.optimize import curve_fit

warnings.filterwarnings('ignore')
```

```python
steps=(200-90)/1100
print(steps)
```

```python
arr = np.arange(90,200,steps)
print(arr)
```

```python
plt.figure(figsize=(10,7),dpi=100)
plt.title('line plot of the array')
plt.plot(arr)
```

```python
len(arr)
```

```python
np.mean(arr)
```

```python
np.std(arr)
```

```python
data=norm.rvs(arr)
```

```python
axis=np.linspace(1,1100,1100)
```

```python
plt.figure(figsize=(10,7),dpi=100)
plt.title('scatter plot')
sns.set_style('darkgrid')
sns.scatterplot(data)
```

```python
def func(x,m,c):
  return m*x+c
```

```python
popt,pcov=curve_fit(func,axis,data,p0=[0,40])
```

```python
popt
```

```python
pcov
```

```python
y_pred = func(axis,popt[0],popt[1])
```

```python
plt.figure(figsize=(10,7),dpi=100)
plt.title('check how good our line fits')
plt.scatter(data,axis)
plt.plot(y_pred,axis,color='r')
```

```python
import yfinance as yahooFinance
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

stock price pridict

```python
google=yahooFinance.Ticker('GOOG')
df=google.history('5Y')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.describe()
```

```python
# overview
sns.set_style('darkgrid')
plt.figure(figsize=(7,8),dpi=190)
plt.title('closing price')
plt.plot(df['Close'])
```

```python
# calculate 50dma
df['fiftyDMA']=df['Close'].rolling(50).mean()
```

```python
# calculate 200dma
df['thDMA']=df['Close'].rolling(200).mean()
```

```python
df.describe()
```

```python
df.drop(columns=['Dividends','Stock Splits'],inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
# plot closing price vs 50dma vs 200dma

sns.set_style('darkgrid')
plt.figure(figsize=[5,6],dpi=200)
plt.title('closing price vs 50dma vs 200dma')
plt.plot(df['Close'],label='Close')
plt.plot(df['fiftyDMA'],label='50DMA')
plt.plot(df['thDMA'],label='200DMA')
plt.legend()
```

infrences
50dma and 200dma are good infrences how the stock is moving

```python
# analyse correlation between each variable
```

```python
plt.figure(figsize=(7,7),dpi=200)
sns.heatmap(df.corr(),annot=True)
```

infrences
there is strong multicollinearity between the variables

```python
# plot distplot of 200dma
sns.set_style('darkgrid')
plt.figure(figsize=(7,5),dpi=200)
plt.title('distplot 200 DMA')
sns.distplot(df['thDMA'])
```
